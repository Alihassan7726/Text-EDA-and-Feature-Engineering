```python
import matplotlib.pyplot as plt
%matplotlib inline
def show(cloud):
    plt.figure(figsize=(40,30))
    plt.imshow(cloud)
    plt.axis('off')
```

```python
from wordcloud import WordCloud, STOPWORDS
# Generate word cloud
wordcloud = WordCloud(width = 2000, height = 1000, random_state=1, background_color='White', 
                      colormap='Set1', collocations=False, stopwords = STOPWORDS).generate(str(df['Text'].tolist()))
# Plot
show(wordcloud)
```

```python
from nltk.stem import WordNetLemmatizer as wnl
import re
def clean_text(text):  
    lower = text.lower()
    words = re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", lower )  
    #pattren = r"[^\w\s\d]"
    #words = re.sub(pattren, "", lower)
    words2 = words.split()
    #print(words2)
    final_words =  [wnl().lemmatize(word) for word in words2 if word not in stopwords.words('english')]
    #return (list(ngrams(final_words, 3)))
    return(final_words)
```

```python

df['Clean_text'] = df['Text'].apply(clean_text)
df
```

```python
wordcloud = WordCloud(width = 2000, height = 1000, random_state=1, background_color='White', 
                      colormap='Set1', collocations=False, stopwords = STOPWORDS).generate(str((df['Clean_text'])))
# Plot
show(wordcloud)
```

```python
from PIL import Image
# Import image to np.array
mask = np.array(Image.open('wo.png'))
# Generate wordcloud
wordcloud = WordCloud(width = 3000, height = 2000, random_state=1, background_color='white', colormap='Set2',
                      collocations=False, stopwords = STOPWORDS, mask=mask).generate(str((df['Clean_text'])))
# Plot
show(wordcloud)
```
